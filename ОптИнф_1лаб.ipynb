```python
#4вариант: sin(pi*alfa*(x-ksi))/pi*(x-ksi)
#[a,b]=[-200,200]
#[p,q]=[-2,2]

import numpy as np  
import matplotlib.pyplot as plt  
import cmath
import math

def K(x, ksi, alfa):
  #alfa = 1
  return math.sin(math.pi*alfa*(x-ksi))/math.pi*(x-ksi)

def f_x(x):
  return np.exp(1j*x/10)
 
def integral(a_b, p_q, alfa):
    
    lena_b = len(a_b)
    lenp_q = len(p_q)
    F_ksi= np.zeros( lena_b, dtype=complex)
    
    h_x = (a_b[lena_b-1] - a_b[0]) / lena_b
    F=0

    for k in range (lenp_q):
      for l in range (lena_b):
        F += (K(a_b[l], p_q[k], alfa))*(f_x(a_b[l]))*h_x
      F_ksi[k] = F
      F = 0         
    return F_ksi

def Plot(x, y):
 # fig = plt.figure()
  plt.plot(x,y)
  plt.grid(True)
  plt.show()
    
a = -200
b = 200
p = -2
q = 2

n = m = 1000

alfa = 1

a_b = np.linspace(a, b, n)
p_q = np.linspace(p, q, m)

Plot(a_b, np.abs(f_x(a_b)))
Plot(a_b, np.angle(f_x(a_b)))

Plot(p_q, np.abs(integral(a_b, p_q, alfa)))
Plot(p_q, np.angle(integral(a_b, p_q, alfa)))
```
